```python
fig = plt.figure(figsize=(10,7))
rows=1
columns= 3

fig.add_subplot(rows,columns,1)
plt.imshow(img_red)
plt.axis("off")
plt.title("RED IMAGE")

fig.add_subplot(rows,columns,2)
plt.imshow(img_blue)
plt.axis("off")
plt.title("BLUE IMAGE")

fig.add_subplot(rows,columns,3)
plt.imshow(img_green)
plt.axis("off")
plt.title("GREEN IMAGE")
```

```python
width, height, channel =300,420,3
img_flag_germany = np.zeros((width ,height, channel),dtype =np.uint8)

img_flag_germany[:100,:]=(0,0,0) 
img_flag_germany[100:200,:]=(255,0,0)
img_flag_germany[200:300,:]=(255,255,0)

plt.imshow(img_flag_germany)
plt.axis("off")
plt.show()
```

```python
width, height, channel =300,420,3
img_flag_russia = np.zeros((width ,height, channel),dtype =np.uint8)

img_flag_russia[:100,:]=(200,200,200) 
img_flag_russia[100:200,:]=(0,0,255)
img_flag_russia[200:300,:]=(255,0,0)

plt.imshow(img_flag_russia)
plt.axis("off")
plt.show()
```

```python
width, height, channel =300,450,3
img_flag_belgium = np.zeros((width ,height, channel),dtype =np.uint8)

img_flag_belgium[:,:150]=(0,0,0) 
img_flag_belgium[:,150:300]=(255,255,0)
img_flag_belgium[:,300:450]=(255,0,0)

plt.imshow(img_flag_belgium)
plt.axis("off")
plt.show()
```

```python
mask_red = (img_flag_russia[:,:,0] == 255) & (img_flag_russia[:,:,1]==0) & (img_flag_russia[:,:,2]==0)
img_flag_russia[mask_red] = [255 ,255,0]
plt.imshow(img_flag_russia)
plt.axis("off")
```

```python
mask_black = (img_flag_belgium[:,:,0] == 0) & (img_flag_belgium[:,:,1]==0) & (img_flag_belgium[:,:,2]==0)
img_flag_belgium[mask_black] = [0 ,0,255]
plt.imshow(img_flag_belgium)
plt.axis("off")
```
